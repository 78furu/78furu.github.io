 ## Git bevezetés, verziókövetés
 
 Git:

- Verziókövető rendszer (Version Control System - VCS) -- mások: Subversion, CVS, and Mercurial
- Több felhasználó munkáját koordinálja
- Változások nyomon követése
- Visszaállítás
- Lokális és távoli repositoryk

[Github](https://github.com):

- Git projektek gyűjtöhelye
- Leginkább nyilt forráskód



### Git célja
- Kód változását követi
- Pillanatképeket (snapshotokat) készít
- Amelyeket <i>kommitálhatunk</i> 
- Majd később megoszthatunk

### Fontosabb fogalmak  és parancsok:
- __Working directory__: _munkakönyvtár,_ ahol dolgozol
- __Staging area (index)__: _készenléti rész_, kész dolgok ideiglenes helye
- __Local repository (history)__: helyi könyvtár, a projekt eddigi életútja tárolva van
- __Remote repository__: távoli könyvtár, felhőben tárolt projekt

~~~bash
git init # start repository
git add # a staging areahoz adja 
git status # miújság a helyzet
git commit # local repohoz adja
git push # remote repohoz
git pull # remote repoból húz le frissítéseket
git clone # egész projektet klónoz le
~~~

## Initialization and 1st commit:

### Project setup

Csináljunk mappát, amiben a projektet tároljuk.

~~~bash
mkdir MyProject
cd MyProject
git init # starting repo
~~~

Hozzáadunk fájlokat:
~~~bash
touch README.md
echo "elso git projektem" >> README.md
cat README.md
touch code.py code2.py
vi code.py # reminder: módosítás: [i], mentés+kilép: [ESC] + :wq
~~~

Nézzük, hogy befolyásolta a gitet

~~~bash
git status # na miujság a helyzet
git add README.md # a staging areahoz . vagy * vagy -a mindent ad
git commit # -m "message" for easy access
~~~

### Track changes

Változtassunk a mappa tartalmán; hozzunk létre, töröljünk, módosítsunk más fájlokat. Nézzük meg a 
~~~bash
git status -s
~~~
paranccsal, mi történt. Bal: staging area, jobb: working directory.

Próba:
~~~bash
git commit -m "my message"
~~~
Error: előbb a staging areahoz

Nézzünk meg a historyt:

~~~bash
git log # --oneline
~~~


Kommitáljunk?

- Ne soronként, de ne is csak a projekt elkészültekor. 
- Ökölszabály: 5-10 alkalommal, egy nap (ha egy projektünk van aznapra). 
- Amint érezzük, hogy "na, működik", tisztítsuk meg, kommitáljuk. 
- Külön commit egy célzott bugfix és egy nem releváns typo. 
- Legyen rövid, egymondatos commit message. 
- (Konvenció a jelenidő: <del>Fixed</del> Fix bug in login.)

### Minek a staging area?

Segítségnek. Csanádok használhatják a
~~~bash
git commit -am "Im sure the bug has flown away."
~~~
rövidítést. De az esetek 90 százalékában inkább két lépésben. Mielőtt biztos jó, azért szeretnénk látni mit módosítottunk, mi a különbség (később).

### Fájlok törlése, átnevezése, áthelyezése

Ha kitöröljük, is hozzá kell adni (tehát tudja, hogy szándékos volt)
~~~bash

git add code2.py
~~~
Egysoros Csanád:
~~~bash

~~~

Hasonló az átnevezés/áthelyezés:

## .gitignore

Olyan fájlokat/foldereket listázunk, amiket nem akarunk, hogy trackeljen.
~~~bash
touch .gitignore
mkdir figs data bin
touch data/results.dat bin/app.o
vi data/results.dat
vi .gitignore
~~~

Miért nem akarunk mindent trackelni?
- Nagy fájlok
- Bináris kódokat
- Eredmények
- Sandboxok
- Logokat
- Privát adatokat
~~~bash
git status -s
git commit -am "added gitignore"
~~~

### Untrack?
__Figyelem:__ a már trackelt fájlokat nem tudjuk _untrackelni_. 

~~~bash
echo hello > bin/app.o
git commit -am "modif and commit app.o accindentaly"
echo bin/ >> .gitignore
git commit -am "add bin/ to gitignore"

echo there >> bin/app.o
git status
~~~

Megoldás: levesszük a staging arearól, hogy ne keresse, ne hívogassa, hagyja békén.

~~~bash
git rm --cached -r bin/ # cached nem törli a working directoryból, -r recursive
git status
~~~

### Remote repo
A <code>git clone</code>, <code>git push</code> és <code>git pull</code> és parancsokkal tudunk kommunikálni a remote repoval.

__Figyelem:__ január óta a fő ág neve nem _master_ hanem _main_. Branchingről később. [További infó az átnevezésről.](https://github.com/github/renaming)

#### Összekapcsolni githubbal:
Előbb hozzá kell adni a paraméterekhez, hogy hol keresse, ez lesz az origin. Majd átnevezhetjük a mainre, végül felküldjük a 
~~~~~~bash
    git push <hova> <melyik agat> 
~~~~~~
pranccsal. Ha az átnevezést kihagyjuk, akkor main => master mindenhol. Az -u flag azért kell, hogy legközelebb, mikor frissíteni akarjuk, tudja honnan szedje le, illetve a statust is innen nézzük.
~~~bash
git remote add origin https://github.com/78furu/myrepo.git # git remote add origin <a te repod címe>
# Opcionális: 
git branch -M main # ez átnevezi a mastert mainre az -M flaggel (felül infó)
git push -u origin main # -u: set upstream for git pull/status
~~~



#### Lehúzni egy egész könyvtárat:

~~~bash
git clone <repo> <hova>
git clone https://github.com/jcjohnson/neural-style .
~~~

#### Megnézzük, hogy a remote reponak mik a tulajdonságai:

~~~bash
git remote -v
git branch -a
~~~

#### Felküldeni a history frissítését

- ellenőrizni a módosításokat:

git diff

git commit -m "fix bug in semi-sideproject"
~~~
- kapcsolatba lépni a repoval:
~~~bash

~~~







## Változások nyomonkövetése (diff)

A különbségek a wd és az index között:
~~~bash
git diff
~~~
index és a history között (mit is akarunk kommitálni):

~~~bash
git diff --staged
~~~

### Értelmezés

Segítség a termináloshoz:

~~~~~~bash
    diff --git a/<file> b/<file>
    index 123abc..21021a 20121
    # ezeket hasonlítjuk össze |^|

    --- a/<file>
    +++ b/<file>
    # jelmagyarázat, legend:|^|
    # a - a régi, a + az új

    @@ -1,4 +1,5 @@
    # chunk header, nem írja ki az egész fájlt |^|
    # a negatívos első sorától négy sort
    # a pozitívos első sorától öt sort ír ki
    
     kozos szoveg
     ez is
     meg ez
    -ez ki lesz törölve (piros)
    +ez uj hozza lett adva (zöld) 
    +meg ez is (zöld)
~~~~~~

A nem termináloshoz használhatunk fancy programokat (de akár a github is jó).

### Nem terminálos programok:

A vscode is jó, kdiff3-at ajánlják:

[Honlap](http://kdiff3.sourceforge.net/): vagy

~~~bash
sudo apt-get install kdiff3
which kdiff3 # másolni
~~~

Beállítás:

~~~bash
git config --global merge.tool kdiff3
git config --global mergetool.kdiff3.path "/usr/bin/kdiff3"
git config --global mergetool.kdiff3.trustExitCode false

git config --global diff.guitool kdiff3
git config --global difftool.kdiff3.path "/usr/bin/kdiff3"
git config --global difftool.kdiff3.trustExitCode false
~~~

Majd újfent terminál
~~~bash
git difftool
~~~

### Visszaállítás, jajj mégse: Ctrl-z-z-z-z-z-z-z...

#### Régi verzió alapján:

Az egyik fájlunkat elrontottunk, hozzuk vissza a régi verzót. Lecsekkoljuk a historyt:
~~~bash
git log --oneline
~~~
Megnézzük, a kiválasztott commit mi is:

~~~bash
git show abc123 # a változások
git show abs123:.gitignore # a végső
~~~
Leveszem az indexről:
~~~bash
git restore --staged basics.py
~~~

#### Töröljünk mindent a legutóbbi commitig

Ehhez egyszerű csak az alábbi parancsot lefuttatni:
~~~bash
git checkout .gitignore
~~~

#### Nagyobb törlések
Elrontotott (legutóbbi) commit messaget az <code>--amend</code> flaggel lehet javítani. Figyelni kell, mert ez a hasht is megváltoztatja. 

~~~bash
git commit --amend -m "modified message"
~~~

Ugyanígy, ha elhagytunk egy fájlt, amit szerettünk volna kommitálni, csak adjuk hozzá az indexhez, majd 
~~~bash
git commit --amend
~~~
__Figyelem:__ Legtöbbször nem akarjuk a git historyt változtatni, mert más felhasználóknak problémákat okozhat.

Ha nagyobb a baj: minden kuka:

~~~bash
git restore . # A nem követett fájlokat nem szedi le
git clean -fd # Vigyázni vele, mert semmiképp nem lehet visszahozni
~~~
-f : force, a nem trackelt fájlokat is kukázza
-d : directory, a mappák tartalmának is helo

Ójaj, kitöröltem (végleg) a helyi gépről:

~~~bash
git rm basics.py
git commit -am "oh i deleteted basics"

git log --oneline
git restore --source=abc123 basics.py # vagy --source=HEAD~1 (eggyel korábbi commit)
~~~


### Kommit törlés helyett revert

Mivel mások már lehet használják a te elrontodd kommitod, célszerű egy plusz kommitot készíteni, ami a visszaállítja egy adott kommit változtatásait:

~~~bash
git log # melyiket
git revert abc123 # commit message előjön
git log # ellenőrzés
~~~

##  Branching 

### Create branch (rügyezik)
Ha valami nagyobb funkciót akarunk beépíteni célszerű új ágat kezdeni, majd azt beépíteni (mergelni) a fő scriptbe. Side projektekhez is hasznos, vagy különböző verziókhoz (pl. fizetős vagy ingyenes app). Tudományos céllal lehet párhuzamosítani. 

~~~bash
git branch parallel # branch for parallelization
git branch # kilistázom
git checkout parallel
git branch # megint kilistázom
~~~


Módosítsuk a szkriptünket, majd commit:
~~~bash
git diff
git add -A
git commit -m "parallelize reading input"
~~~

Ez nem módosította sem a local, sem a remote maint. Viszont szeretnénk, hogy a branch megjelenjen a remote repoban is:

~~~bash
git push -u origin parallel # tehát az ottani parallel ág az itteni
git branch -a
~~~

### Végeztünk a sideprojekttel
#### Beépítjük (merge)

Ha végeztünk, mielőtt célszerű frissíteni a main ágat is.

Lépések:
- visszamegyünk a main ágra
- frissítjük a main ágat 
- (kiírjuk a merged ágakat)
- beépítjük a parallel ágat
- frissítjük a remote repot

~~~bash
git checkout main
git pull origin master
git branch --merged
git merge parallel
git push origin master
~~~

#### Kitöröljük (delete)
Ha végeztünk, és nem akarunk több módosítást csinálni az adott ágon, célszerű kitörölni.

~~~bash
git branch --merged
git branch -d parallel
git branch -a
git push origin --delete parallel
~~~

### Branchek közötti mozgás/mozgatás

#### Kommitot viszünk
Ajjajj, nem rossz branchre commitoltam. Itt a megoldás: átmásolom a jó ágra a kívánt snapshotot (kommitot) a <code>git cherry-pick</code> paranccsal, majd törlöm a rossz ágról a <code>git restore</code> paranccsal.

Hiba:

~~~bash
git branch docs
git checkout main
echo "# basic python funcs" >> basics.py
git commit -am "create docs"
~~~

Megkeressük a hasht:
~~~bash
git log # hash első 6-7 karaktere crtl+c
git checkout docs # váltás
git log # nincs ott az a kommit
git cherry-pick abc123 # áthúzzuk
git log # ott van
~~~

Kitöröljük a rossz kommitot. Oké, de a wd nem változik? 

- --soft: nem bántja a wd-t, de törli az index tartalmát
- --mixed: DEFAULT, meghagyja az index tartalmát is
- --hard: törli a wd-ban lévő változtatásokat is, amennyiben követte őket

~~~bash
git restore --hard abc123
git clean -fd # törli a nem követetteket is
~~~


#### Változtatásokat viszünk

Stashsel át tudjuk vinni a nem kommitált változtatásokat. Erről később.

### Végleges törlés visszavonása

Ha véletlenül mégis --harddal töröltünk, van megoldás, ugyanis a git garbage collectora (kb) 30 napig él. Nézzük, hogy mit hivatkoztunk utoljára:

~~~bash
git reflog
git checkout abc123
~~~

Egy detached branchre visz, de a fájlokat vissza tudjuk kapni.

### Félretenni változtatásokat (stash)

Ha úgy alakul, hogy még nincs ideje a commitnak, akkor tudjuk stashelni. Ekkor az adott branchen félrerakjuk, és majd legközelebb elővesszük.

~~~bash
git checkout docs
echo "it is very important to calcu" >> any.txt
git stash save "start larger docs"
git stash list
~~~

Ezután félre van rakva (git diffel/statussalmegnézhetjük). Mikor újra kell, elővehetjük:

~~~bash
git stash apply stash{0} # ez nem törli a félrerakottat
git checkout -- . # vissza
git stash pop # a legelsőt visszarakja és eldobja
~~~

Mikor jó ez? Dolgozol például egy párhuzamosításon vagy docstringelésen, de a főnök szól, hogy mi ez a kimeneti formátum. Aktuálisat stashre, megcsinálni a formázást, vissza a sajáthoz. 

Ha el akarjuk dobni, drop:
~~~bash
git stash drop stash{0} # egy adott
git stach clear # mindent. óvatosan ezzel
~~~

### Merge conflict

Ha egy másik ágon már kommitáltunk egy változtatást, és a git nem tudja a konfliktust feloldani, akkor jön a merge conflict. 

~~~bash
git branch b
git checkout b
echo "on branch b" >> basics.py
git commit -am "added branch text"
git checkout main
echo "on branch a" >> basics.py
git merge b
~~~
Errort dob:
~~~~~~bash
    error: Your local changes to the following files would be overwritten by merge:
	basics.py
    Please commit your changes or stash them before you merge.
    Aborting
~~~~~~


Ha kommitáljuk:
~~~bash
git add *
git commit -m 'comm to merge conf on master'
git merge b
~~~
Ekkor merge conflict:

~~~~~~bash
    Auto-merging basics.py
    CONFLICT (content): Merge conflict in basics.py
    Automatic merge failed; fix conflicts and then commit the result.
~~~~~~
