# <center> About Causality </center>

<img style="float: right;" src="pics/causality_types.png" alt = "caus" width = 200>


* Having two discrete time series $X$ and $Y$
* One would like to infer the relation between them
* Types of connections:
     - Independent $X \perp Y$ 
     - Unidirectional $X\rightarrow Y$ or $X\leftarrow Y$
     - Bidirectional $X \leftrightarrows Y$ 
     - (Hidden) Common cause $X \Lsh  \Rsh  Y $

```python
from scipy.stats import pearsonr, entropy
from scipy.signal import correlate, coherence
import scipy.signal as signal
from __future__ import print_function
from sklearn.feature_selection.mutual_info_ import mutual_info_score
import os.path
#from mne import connectivity
```

```python
 Fs, f = 8000, 5
x = np.arange(Fs)
A = np.sin(2 * np.pi * f * x / Fs)
B = np.sin(2 * np.pi * f * x / Fs+ pi/6)
C = np.sin(2 * np.pi * f * x / Fs+ pi/2)
D = [0.1 for i in x]
x = np.linspace(0,1,Fs)
```

```python
def entropy1(labels, base=None):
    value,counts = np.unique(labels, return_counts=True)
    return entropy(counts, base=base)
```

####  <center> Simpler connections in time domain </center>
- Pearson correlation (non-directed, model-based):
$$
r_{xy}=\frac {\sum _{i=1}^{n}(x_{i}-{\bar {x}})(y_{i}-{\bar {y}})}{{\sqrt {\sum _{i=1}^{n}(x_{i}-{\bar {x}})^{2}}}{\sqrt {\sum _{i=1}^{n}(y_{i}-{\bar {y}})^{2}}}}
    $$
    

```python
print("Pearson correlation| A-B:", round(pearsonr(A,B)[0],4), "A-C:",
      round(pearsonr(A,C)[0],4), "A-D:", round(pearsonr(A,D)[0],4))
```

- Mutual information (non-directed, model-free):
$$
I(X;Y) = \sum_{x,y} P_{XY}(x,y) \log
{P_{XY}(x,y) \over P_X(x) P_Y(y)}
= E_{P_{XY}} \log{P_{XY} \over P_X P_Y}
$$

```python
print("The entropies|       A: ", round(entropy1(A),4), " B: ",
      round(entropy1(B),4), " C: ", round(entropy1(C),4), " D: ", round(entropy1(D),4))
print("Mutual information| A-B:", round(mutual_info_score(A, B),4), "A-C:",
      round(mutual_info_score(A,C),4), "A-D:", round(mutual_info_score(A,D),4))
print("As a reference|     A-A:", round(mutual_info_score(A,A),4), "B-B:",
      round(mutual_info_score(B,B),4), "C-C:", round(mutual_info_score(C,C),4))
```

- Cross correlation (directed, model-based):
    
$$
(f\star g)[n] =\sum _{m=-\infty }^{\infty }{f[m]^*}g[m+n]
$$

```python
figsize(8,5)
```

```python
grid()
plot(correlate(A,B))
plot(correlate(A,C))
plot(correlate(A,D))
legend(["A-B", "A-C", "A-D" ], loc = 'best')
xlabel('Data point');
```

# <center> Frequency domain </center>

- Coherence
$$
 \gamma _{xy}^{2}(f)={\frac {|S_{xy}(f)|^{2}}{S_{xx}(f)S_{yy}(f)}}
$$


```python
Fs, f1, f2 = 8000, 1000, 1500
x = np.arange(Fs)
A = np.sin(2 * np.pi * f1 * x / Fs)
B = A + np.sin(2 * np.pi * f2 * x / Fs+ pi/6)
B = B/2
x = np.linspace(0,1,Fs)
noise_power = 0.001 * Fs / 2
B = B+ np.random.normal(scale=0.1*np.sqrt(noise_power), size=Fs)
```

```python
f, c = coherence(A,B, fs = Fs )
```

```python
plt.plot(x, array([A,B]).T)
plt.xlim(0.4, 0.42)
plt.grid()
plt.xlabel("Time [sec]")
plt.ylabel("Signal")
plt.legend(["A (1kHz)", "A + B (1.5kHz) + noise"], loc = "best");
```

```python
plt.semilogy(f,c)
plt.grid()
plt.xlabel("Frequency [Hz]")
plt.ylabel("Coherence");
```

# <center> Granger causality </center>

<center> Linear autoregression models: </center>
<br>
$$
    x\left[t\right] = \sum^m_{i=1} a_i x\left[t-i \right] + \epsilon_1 \text{ and } y\left[t\right] = \sum^m_{i=1} b_i y\left[t-i \right] + \epsilon_2
$$

<br>
<center>  Taking into account the information from the other signal  </center>
<br>
$$
    x\left[t\right] = \sum^m_{i=1} c_i x\left[t-i \right] + \sum^q_{i=1} d_i y\left[t-i \right] + \epsilon_3 
$$
$$y\left[t\right] = \sum^m_{i=1} e_i x\left[t-i \right] + \sum^q_{i=1} f_i y\left[t-i \right] + \epsilon_4
$$
<br>

----
<p style="font-size:80%;" align = "bottom"> $\textit{Kristóf Furuglyás, Theoretical Physics Seminar, 2019 Fall }$ </p>

## <center> Granger Causality</center>
<br>
<center> In this sense, causality is provided by: </center>
<br>
$$
F_{y\rightarrow x} = \ln \left(\frac{var\left( \epsilon_1\right)}{var\left( \epsilon_3\right)}  \right) \text{ and } F_{x\rightarrow y} = \ln \left(\frac{var\left( \epsilon_2\right)}{var\left( \epsilon_4\right)}  \right) 
$$
<br>
<center> Evaluation F test: </center>
<br>
$$ 
 F_{y\rightarrow x} = \ln \left(\frac{\frac{var\left( \epsilon_1\right) - var\left( \epsilon_3\right)}{m}}{\frac{var\left( \epsilon_3\right)}{T-2m-1}}  \right) 
$$
<br>
<center><b> Ref: Granger, Clive WJ., 1969 </b></center>

----
<p style="font-size:80%;" align = "bottom"> $\textit{Kristóf Furuglyás, Theoretical Physics Seminar, 2019 Fall }$ </p>

### <center> Problems with Granger causality </center>

As George Sugihara mentioned (Sugihara et al. (2012)):

> However, as Granger realized early on, this approach may be problematic in deterministic settings, especially in dynamic systems with weak to moderate coupling. [...]

> That is to say, in deterministic dynamic systems (even noisy ones), if X is a cause for Y, information about X will be redundantly present in Y itself and cannot formally be removed from U—a consequence of Takens’ theorem.

#  <center> Dimensional causality </center>

<ul>
    <li> Method developed by Benkő et al. (2018) to detect the hidden common cause based on the dimensions of the fractals created in the joint space. </li>
    <li> <b>Key point: the cause does not increases the dimension of the consequence in the joint space, the information is already there! </b></li>
    </ul>
    


<center><img style="float:middle" src="pics/dc_basics.png" alt = "tdccm" width = 600></center>



----
<p style="font-size:80%;" align = "bottom"> $\textit{Kristóf Furuglyás, Theoretical Physics Seminar, 2019 Fall }$ </p>

## <center> Mathematical approach </center>

Information dimension introduced by Rényi (1959):
$$
    d_X = \lim_{N\rightarrow\infty} \frac{1}{\log N}  H\left( \left[X\right]_N\right),
$$

Since $X$ is discrete and finite, if the variable lives in a $D$ dimensional space, a proper estimate needs at least $10^D-30^D$ sample points. An approximate form :

$$
    d_{X, 1/N} = \frac{1}{\log N} H\left( \left[X\right]_N\right).
$$


Assumption 1: time series are stationary. Local intrinsic dimenson:

$$
D_X\left(x\right) = \lim_{r\rightarrow 0} \frac{1}{\log r} \log P \left( X \in B\left(x,r \right)\right),
$$

The intrinsic dimension:

$$
d_X = D_X = E\left( D_X\left( X \right) \right)
$$


----
<p style="font-size:80%;" align = "bottom"> $\textit{Kristóf Furuglyás, Theoretical Physics Seminar, 2019 Fall }$ </p>


## <center> Mathematical approach </center>

Assumption 2: the embedded manifolds are homogeneous with respect to (the existing) dimension. Local intrinsic dimension estimates:

$$
\hat{D} \left(x \right)_r = \frac{1}{\log r} \log \left| N \left(x,r \right) \right|.
$$

Global dimension estimate:

$$
\hat{D}_{X,r} = \frac{1}{n} \sum_{i=1}^n \hat{D} \left(x_i \right)_r
 $$
 
 

For the $k$-th nearest neighbour, one shall calculate the dimension from the distance $r(x,k) = d(x, X^k(x))$. In this setting, the resolution is:

$$
r^D \approx \frac{k}{n}
$$

## <center> Detection of causality </center>

 - Independent 
     $$X \perp Y \Rightarrow D_J = D_X + D_Y$$
     <br>
 - Unidirectional 
     $$X\rightarrow Y \Rightarrow D_J = D_Y <D_X + D_Y $$
      <br>
 - Bidirectional 
     $$X \leftrightarrows Y \Rightarrow  D_J = D_X = D_Y$$
      <br>
 - (Hidden) Common cause 
     $$X \Lsh  \Rsh  Y \Rightarrow \max\left(D_X, D_Y \right) < D_J < D_X + D_Y$$
